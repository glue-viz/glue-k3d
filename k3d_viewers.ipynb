```python
from glue.config import settings
from glue.core import Data
from glue_jupyter import jglue
from glue_k3d.viewers.scatter.viewer import K3DScatterView
from glue_k3d.viewers.volume.viewer import K3DVolumeView
from matplotlib import colormaps
from random import randint
import numpy as np
```

```python
app = jglue()
```

```python
N = 100
x = [randint(0, N) for _ in range(N)]
y = [randint(0, N) for _ in range(N)]
z = [randint(0, N) for _ in range(N)]
scatter_data = Data(label="Scatter Data", x=x, y=y, z=z)
scatter_data.style.color = "#ff0000"
app.data_collection.append(scatter_data)
```

```python
viewer = app.new_data_viewer(K3DScatterView, data=scatter_data)
```

```python
t = np.arange(24).reshape((4, 3, 2))
volume_data = Data(label="Volume Data", t=t)
app.data_collection.append(volume_data)
```

```python
volume = app.new_data_viewer(K3DVolumeView, data=volume_data)
layer = volume.layers[0]
layer.state.cmap_mode = "Linear"
layer.state.alpha = 0.5
layer.state.cmap = colormaps.get_cmap("plasma")
```
